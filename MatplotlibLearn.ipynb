```python
try:
    f = open("diamond.dat")
    #read a line to get rid of the first line in order to 
    header = f.readline().rstrip().split(',')
    print(header)
    headerLength = len(header)
    datImport = np.loadtxt(f, delimiter=',', usecols=range(1,headerLength))

finally:
    f.close()
```

```python
x = datImport[:,5]
bands = []
for i in range(5, headerLength - 1):
    bands.append(datImport[:,i])
```

```python
figure = plt.figure()
for band in bands:
    plt.plot(x, band, '.-')
    print(band)
```
