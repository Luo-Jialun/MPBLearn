```python
import math
import meep as mp
from meep import mpb
from scipy.optimize import minimize_scalar
```

```python
def SqaureLatticeTMGap(r):
    num_bands = 8

    boundary_k_points = [mp.Vector3(),
            mp.Vector3(0.5),
            mp.Vector3(0.5, 0.5),
            mp.Vector3()]

    k_points = mp.interpolate(4, boundary_k_points)

#     for k_vec in k_points:
#         print(k_vec)

    geometry = [mp.Cylinder(radius=0.3, material=mp.Medium(epsilon_diag=mp.Vector3(1, 1, 12)))]
    geometry_lattice = mp.Lattice(size=mp.Vector3(1,1))
    resolution = 32
    

    solver = mpb.ModeSolver(num_bands=num_bands,
                        k_points=k_points,
                        geometry=geometry,
                        geometry_lattice=geometry_lattice,
                        resolution=resolution)
    
    solver.num_bands = 8
    solver.mesh_size = 3
    solver.geometry = [mp.Cylinder(radius=0.3, material=mp.Medium(epsilon_diag=mp.Vector3(1, 1, 12)))]
    solver.default_material = mp.Medium(epsilon_diag=mp.Vector3(12, 12, 1))

    
    print(solver.num_bands)
    solver.run()
    
    return -1 * solver.retrieve_gap(1)

    
```

```python
SqaureLatticeTMGap(0.3)
```
